Actuarially fair contracts

$$
P = \int_0^{\underline{PV}} (\underline{PV}-x) \cdot f(x) \, dx
$$

Example:

Suppose $x$ is uniformly distributed between 0 and 8. Then $f(x) = 1/9$.

Assume that the trigger is set at $\underline{PV} = 6$. Then

$$
P = \int_0^{\underline{6}} (6-x) \cdot f(x) \, dx = \frac{1}{9} \cdot 
\left. \left (6x-\frac{x^2}{2} \right)  \right|_0^6  = \frac{1}{9} \cdot (36-18) = 2
$$

This is as depicted above.  The insured pays a \$2 premium following any outcome $x$.  The insurer pays out $6-x$ if $x \leq 6$.  The expected premium (collected in every state is \$2) which is exactly equal to what the insurer pays out.
